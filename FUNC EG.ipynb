```python
#ENCAPSULATION EG - 1
class MobilePhone:
    def __init__(self, model, pin):
        self.model = model
        self.__pin = None
        if 1000 <= pin <= 9999:
            self.__pin = pin
        print(f"{self.model} initialized.")
    def set_pin(self, new_pin):
        if 1000 <= new_pin <= 9999:
            self.__pin = new_pin
            print("PIN updated successfully.")
        else:
            print("Invalid PIN. Must be a 4-digit number (1000-9999).")
    def check_pin(self, entered_pin):
        if entered_pin == self.__pin:
            print("Correct PIN. Phone unlocked!")
        else:
            print("Incorrect PIN. Access denied.")
phone = MobilePhone("Samsung Galaxy", 4821)
phone.check_pin(1234)
phone.check_pin(4821)
phone.set_pin(99)
phone.set_pin(7654)
phone.check_pin(4821)
phone.check_pin(7654)   
```

```python
#INHERITANCE EG - 1
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def introduce(self):
        print(f"Hi, I am {self.name} and I am {self.age} years old.")
class Student(Person):
    def __init__(self, name, age, student_id, grade):
        super().__init__(name, age)
        self.student_id = student_id
        self.grade = grade
    def introduce(self):
        super().introduce()
        print(f"Student ID : {self.student_id}")
        print(f"Grade      : {self.grade}")
class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject
    def introduce(self):
        super().introduce()
        print(f"Subject    : {self.subject}")
    def take_class(self, topic):
        print(f"{self.name} is teaching {topic} today.")
s = Student("Arun", 20, "S2045", "A")
t = Teacher("Mrs. Kavitha", 35, "Mathematics")
s.introduce()
print()
t.introduce()
t.take_class("Quadratic Equations")
```
